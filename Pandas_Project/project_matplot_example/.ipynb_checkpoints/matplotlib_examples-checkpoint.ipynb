## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Gas Price Data

## Line Graph

```python
gas  = pd.read_csv('gas_prices.csv', index_col = 'Year')
gas.head()
```

```python
plt.figure(figsize=(12,7),dpi=300)

plt.title('Gas Prices over Time (in USD)', fontdict={
    'fontweight' : 'bold',
    'fontsize' : 18
})
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.plot(gas,label=gas.columns, marker = '.', markeredgecolor='k')
plt.xticks(gas.index, rotation=45)
plt.legend()

plt.savefig('Gas_price_figure.png')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)')

plt.plot(gas.index, gas.USA, 'b.-')
plt.plot(gas.index, gas.Canada, 'r.-')
plt.plot(gas.index, gas['South Korea'], 'g.-')
plt.plot(gas.index, gas.Australia, 'y.-')

plt.xticks(gas.index[::3])
plt.xlabel('Year')
plt.ylabel('US DOLLARS')

plt.legend()

plt.show()
```

# FIFA Data

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

## Histograms

```python
bins = list(np.arange(30,110,10))

plt.hist(fifa.Overall, bins=bins, color = '#abcdef')

plt.xticks(bins)
plt.ylabel('Number of Player')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

## Pie Chart #1

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels=labels, colors=colors, autopct='%0.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

## Pie Chart #2

```python
plt.figure(figsize=(8,5),dpi=100)
plt.style.use('fast')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','over 200']
explode = (0.4,0,0,0,0.4)

plt.title('Weights of Professional Football Players (lbs)')

plt.pie(weights, labels=labels, explode=explode, pctdistance=0.8, autopct='5.2f %%')
plt.show()
```

## Box and Whiskers Chart

```python
plt.figure(figsize=(6,8), dpi=100)

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

bp = plt.boxplot([barcelona,madrid,revs], labels=['FC Barcelona', 'Real Madrid', 'NE Revolution'], patch_artist=True,medianprops={
    'linewidth':2
},boxprops={
    'color' : '#4286f4',
    'linewidth' : 2,
    'facecolor' : '#e0e0e0'
})

plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
