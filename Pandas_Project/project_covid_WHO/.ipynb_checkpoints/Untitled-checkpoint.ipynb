## PROJECT COVID-19 ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vaccine = pd.read_csv('vaccination-data.csv')
covid = pd.read_csv('WHO-COVID-19-global-data.csv')
```

```python
who_region = list(covid.groupby('WHO_region').count().index)
who_region
```

```python
total_cases = pd.DataFrame(covid.groupby('WHO_region').New_cases.sum())
total_cases
```

```python
plt.figure(figsize=(12,6))
labels = ['African Region', 'Region of the Americas', 'Eastern Mediterranean Region', 'European Region', 'Other', 'South-East Asian Region', 'Western Pacific Region']
plt.pie(total_cases.New_cases,labels=labels, autopct='%0.2f %%', pctdistance=0.8)

plt.title('Cases in Different WHO Regions')

plt.show()
```

```python
country_cases = pd.DataFrame(covid.groupby('Country').New_cases.sum())
country_cases
```

```python
plt.figure(figsize=(12,5))
plt.xticks(country_cases.index,rotation=60)
plt.bar(country_cases.index,country_cases.New_cases)
```
